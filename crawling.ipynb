```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime

# 현재 시간을 가져오기
current_time = datetime.now()

# 날짜, 시간, 분 등 추출
formatted_date = current_time.strftime('%Y-%m-%d')
formatted_time = current_time.strftime('%H:%M:%S')

prefix = "https://kin.naver.com"

collected_prompt = []
collected_label = []

page_per_question = 20
loop = 10
for i in range(loop):
    # URL 생성
    print(f'################# {i*page_per_question}/{loop*page_per_question} ################')

    url = f'{prefix}/qna/kinupList.naver?queryTime={formatted_date}%20{formatted_time}&page={i}'
    print("생성된 URL:", url)
    print("\n\n")

    # 네이버에게 웹 크롤링으로 접근하는 것에 대한 정책을 준수하기 위해 User-Agent 설정
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    # td 태그 중 class가 "title"인 것을 찾아서 처리
    for td in soup.find_all('td', class_='title'):
        a_tag = td.find('a')
        if a_tag:
            link = a_tag['href']
            link = prefix + link
            print("link: ", link)
            # 찾아낸 링크로 접속하여 페이지의 내용을 가져옴
            page_response = requests.get(link, headers=headers)
            page_soup = BeautifulSoup(page_response.text, 'html.parser')

            has_prompt = False
            has_answer = False

            # div 태그 중 class가 "c-heading__content"인 것을 찾아서 출력
            for div_tag in page_soup.find_all('div', class_='c-heading__content'):
                has_prompt=True
                prompt = div_tag.get_text().strip()
                print("prompt: ", prompt)


            answer = ""
            # p 태그 중 class가 "se-text-paragraph"인 것을 찾아서 출력
            for p_tag in page_soup.find_all('p', class_='se-text-paragraph'):
                if has_answer == False:
                    has_answer=True
                answer += p_tag.get_text()


            print("answer: ", answer)
            print("has_prompt: ", has_prompt)
            print("has_answer: ", has_answer)

            if has_prompt and has_answer:
                collected_prompt.append(prompt)
                collected_label.append(answer)

            print("\n\n")
            has_prompt = False
            has_answer = False

    print("\n\n")

print("수집한 프롬프트: ", collected_prompt)
print("수집한 프롬프트 수: ", len(collected_prompt))
print("수집한 라벨: ", collected_label)
print("수집한 라벨 수: ", len(collected_label))
```
