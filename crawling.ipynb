```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime

def get_current_datetime():
    return datetime.now()

def format_datetime(dt):
    return dt.strftime('%Y-%m-%d %H:%M:%S')

def make_request(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        return response.text
    except requests.exceptions.RequestException as e:
        print(f"Error making request: {e}")
        return None

def scrape_page(link):
    page_response = make_request(link)
    if not page_response:
        return None, None

    page_soup = BeautifulSoup(page_response, 'html.parser')

    prompt = page_soup.find('div', class_='c-heading__content')
    answer = '\n'.join(p_tag.get_text() for p_tag in page_soup.find_all('p', class_='se-text-paragraph'))

    return prompt.get_text().strip() if prompt else None, answer

def main():
    prefix = "https://kin.naver.com"
    collected_prompt = []
    collected_label = []

    page_per_question = 20
    loop = 10

    current_time = get_current_datetime()
    formatted_date_time = format_datetime(current_time)

    for i in range(loop):
        print(f'################# {i * page_per_question}/{loop * page_per_question} ################')
        url = f'{prefix}/qna/kinupList.naver?queryTime={formatted_date_time}&page={i}'
        print("생성된 URL:", url)

        page_content = make_request(url)
        if not page_content:
            continue

        soup = BeautifulSoup(page_content, 'html.parser')

        for td in soup.find_all('td', class_='title'):
            a_tag = td.find('a')
            if a_tag:
                link = prefix + a_tag['href']
                print("link: ", link)

                prompt, answer = scrape_page(link)

                if prompt and answer:
                    collected_prompt.append(prompt)
                    collected_label.append(answer)

                print("\n\n")

    print("수집한 프롬프트: ", collected_prompt)
    print("수집한 프롬프트 수: ", len(collected_prompt))
    print("수집한 라벨: ", collected_label)
    print("수집한 라벨 수: ", len(collected_label))

if __name__ == "__main__":
    main()
```
