```python
import requests
from bs4 import BeautifulSoup

prefix = "https://kin.naver.com"
url = "https://kin.naver.com/qna/kinupList.naver?queryTime=2023-12-03%2009%3A58%3A10&page=2"

# 네이버에게 웹 크롤링으로 접근하는 것에 대한 정책을 준수하기 위해 User-Agent 설정
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# td 태그 중 class가 "title"인 것을 찾아서 처리
for td in soup.find_all('td', class_='title'):
    a_tag = td.find('a')
    if a_tag:
        link = a_tag['href']
        link = prefix+link
        print("link: ", link)
        # 찾아낸 링크로 접속하여 페이지의 내용을 가져옴
        page_response = requests.get(link, headers=headers)
        page_soup = BeautifulSoup(page_response.text, 'html.parser')

         # div 태그 중 class가 "c-heading__content"인 것을 찾아서 출력
        for div_tag in page_soup.find_all('div', class_='c-heading__content'):
            prompt = div_tag.get_text().strip()
            print("prompt: ", prompt)

        answer=""
         # p 태그 중 class가 "se-text-paragraph"인 것을 찾아서 출력
        for p_tag in page_soup.find_all('p', class_='se-text-paragraph'):
            answer += p_tag.get_text()

        print("answer: ", answer)
        print("\n\n")

```
